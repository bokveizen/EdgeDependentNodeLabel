```python
import pandas as pd
import os
import numpy as np
import copy
import random
```

```python
dataname = "DBLP2"
k = "10000"
```

```python
a = pd.read_csv("../dataset/" + dataname + "/whole_data_" + k + ".txt")
a
```

```python
tmp = list(a["node reindex"].unique())
```

```python
new_tmp = copy.deepcopy(tmp)
random.shuffle(new_tmp)
```

```python
mapping = {}
for i,j in zip(tmp, new_tmp):
    mapping[i] = j
```

```python
a = a.replace({"node reindex": mapping})
a
```

```python
a.to_csv("../dataset/" + dataname + "/whole_data_" + k + ".txt", index=False)
```
