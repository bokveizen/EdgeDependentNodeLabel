```python
import pandas as pd
import os
import numpy as np
```

```python
dataname = "emailEnron"
k = "0"
```

```python
usecols = ["degree", "eigenvec", "kcore", "pagerank", 
           "degree_rank", "eigenvec_rank", "kcore_rank", "pagerank_rank", 
           "degree_avg", "degree_max", "degree_min", "degree_sum", 
           "eigenvec_avg", "eigenvec_max", "eigenvec_min", "eigenvec_sum",
           "kcore_avg", "kcore_max", "kcore_min", "kcore_sum",
           "pagerank_avg", "pagerank_max", "pagerank_min", "pagerank_sum",
           "pos"]
```

```python
a = pd.read_csv("dataset/" + dataname + "/whole_data_" + k + ".txt", usecols=usecols)
```

```python
a
```

```python
pos0 = np.zeros(len(a))
pos1 = np.zeros(len(a))
pos2 = np.zeros(len(a))
for i, row in a.iterrows():
    if row['pos'] == 0:
        pos0[i] = 1
    elif row['pos'] == 1:
        pos1[i] = 1
    elif row['pos'] == 2:
        pos2[i] = 1
```

```python
print(pos0)
```

```python
a['pos0'] = pos0
a['pos1'] = pos1
a['pos2'] = pos2
```

```python
a = a.drop(columns='pos')
```

```python
a
```

```python
# plt.scatter(a['degree'], a['pos0'])
```

```python
cor = a.corr()
```

```python
cor
```

```python
cor['pos'] = cor[['pos0', 'pos1', 'pos2']].mean(axis=1)
```

```python
cor
```

```python
whole = cor.to_numpy()
# print(whole)
```

```python
sample = cor.to_numpy()
# print(sample)
```

```python
whole = (whole - np.min(whole)) / (np.max(whole) - np.min(whole))
```

```python
sample = (sample - np.min(sample)) / (np.max(sample) - np.min(sample))
```

```python
diff = np.power(whole - sample, 2)
# print(diff)
print(np.mean(diff))
```
