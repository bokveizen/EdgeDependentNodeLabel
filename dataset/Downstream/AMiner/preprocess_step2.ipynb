```python
inputdir  = "./"
```

```python
numhedges = 0
numnodes = 0
hedge2node = []
hedge2nodepos = []
hedge2nodepos_predict = []
node2hedge = []

sampled_paperid = set()
paperid2hindex = {}
paperid2category = {}
hindex2paperid = {}

authorid2vindex = {}
vindex2authorid = {}

category_indexing = {}

# Read Data ======================================================================================================================
# hypergraph --------------------------------------------------------------------
with open(inputdir + "sampled_paperid.txt", "r") as f:
    for line in f.readlines():
        paperid, category = line.rstrip().split("\t")
        if category not in category_indexing:
            category_indexing[category] = len(category_indexing)
        paperid2category[paperid] = category
        sampled_paperid.add(paperid)

with open(inputdir + "hypergraph.txt", "r") as f, open(inputdir + "hypergraph_pos.txt", "r") as pf:
    for line, pline in zip(f.readlines(), pf.readlines()):
        nodes = line.rstrip().split("\t")
        node_poses = pline.rstrip().split("\t")[1:]
        paperid = nodes[0]
        nodes = nodes[1:]
        if len(nodes) == 1:
            continue
        elif paperid not in sampled_paperid:
            continue
        hindex = numhedges
        paperid2hindex[paperid] = hindex
        hindex2paperid[hindex] = paperid
        numhedges += 1
        hedge2node.append([])
        hedge2nodepos.append([])
        hedge2nodepos_predict.append([])
        
        for authorid, _vpos in zip(nodes, node_poses):
            if authorid not in authorid2vindex:
                vindex = numnodes
                authorid2vindex[authorid] = vindex
                vindex2authorid[vindex] = authorid
                numnodes += 1
                node2hedge.append([])
            vindex = authorid2vindex[authorid]
            if _vpos == 1:
                vpos = 2
            elif _vpos == len(nodes):
                vpos = 2
            else:
                vpos = 1
            hedge2node[hindex].append(int(vindex))
            hedge2nodepos[hindex].append(int(vpos))
            node2hedge[vindex].append(hindex)
```

```python
portion =  ['Computer Graphics','Computational Linguistics', 'Human Computer Interaction','Databases & Information Systems']
```

```python
connected_component = []
check = [False for _ in range(numhedges)]

for h in range(numhedges):
    paperid = hindex2paperid[h]
    if paperid2category[paperid] not in portion:
        continue
    if check[h] is False:
        queue = [h]
        component = []
        while len(queue) > 0:
            cur = queue.pop(0)
            check[cur] = True
            component.append(cur)
            for v in hedge2node[cur]:
                for nh in node2hedge[v]:
                    npaperid = hindex2paperid[nh]
                    if paperid2category[npaperid] not in portion:
                        continue
                    if check[nh] is False and nh not in queue:
                        queue.append(nh)
        connected_component.append(component)
        print(len(component))
```

```python
component = connected_component[0]
print(len(component))
```

```python
from collections import defaultdict
```

```python
category_dist = defaultdict(int)
for hindex in component[:10000]:
    paperid = hindex2paperid[hindex]
    category = paperid2category[paperid]
    category_dist[category] += 1
print(category_dist)
```

```python
with open("sampled_paperid_10000.txt", "w") as f:
    for hindex in component[:10000]:
        paperid = hindex2paperid[hindex]
        
        f.write(str(paperid) + "\n")
```
